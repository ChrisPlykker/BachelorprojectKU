```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Font selecter :
plt.rc('font',family='Times New Roman')
```

```python
#This part is for figure 6.
def approx_square_well_energies(x):
    """ 
    
    Returns the variational approximation to the energies for the infinite square well at the array of given x-values(x)
    
    """
    return np.pi**2*((x + 2)**2-8/3)
def square_well_energies(x):
    """ 
    
    Returns the exact energies for the infinite square well at the array of given x-values(x)
    
    """
    return np.pi**2*(x+1)**2
def square_well_energies_step_ahead(x):
    """ 
    
    Returns the exact energies for the infinite square well, but with n starting at 1, at the array of given x-values(x)
    
    """
    return np.pi**2*(x+2)**2
line = np.linspace(0, 8, 9)
approx_square_well_y_vals = approx_square_well_energies(line)
square_well_energies_y_vals = square_well_energies(line)
square_well_energies_y_vals_step_ahead = square_well_energies_step_ahead(line)
```

```python
#Plot of the above equations
#plt.title("Comparison of approximate and exact ISW energies")
plt.xlabel("n")
plt.ylabel("E(n)")
plt.plot(line, approx_square_well_y_vals, ".", label = "Approximate energi values")
plt.plot(line, square_well_energies_y_vals, ".", label = "Exact energy values")
plt.plot(line, square_well_energies_y_vals_step_ahead, ".", label = "Exact energy values, but starting one n ahead")
plt.xlim(0, 9)
plt.legend()
plt.grid()
plt.savefig("ISW comparison.png")
```

```python
#This part is for figure 1

L = 1

def ISW_potential(x, L):
    """ 
    
    Returns infinite square wells potential at the given x-values(within the well)
    
    """
    return -np.pi**2/L**2*x/x#x/x is just a weird technicality to make sure it actually returns more than one point...

def ISW_potential_partner(x, L):
    """ 
    
    Returns infinite square wells partner potential at the given x-values(within the well)
    
    """
    return np.pi**2/L**2 *(2/(np.sin(np.pi*x/L))**2 - 1)

ISW_line = np.linspace(0.1, L-0.1, 1000)
ISW_potential_vals = ISW_potential(ISW_line, L)
ISW_potential_partner_vals = ISW_potential_partner(ISW_line, L)

```

```python
#plt.title("Infinite Square Well Potential and Partner Potential Comparison")
plt.xlabel("x")
plt.ylabel("V(x)")
plt.plot(ISW_line, ISW_potential_vals, label = "Infinite Square Well potential")
plt.plot(ISW_line, ISW_potential_partner_vals, label = "Infinite Square Well Partner Potential")
plt.xlim(0.1, L-0.1)
plt.legend()
plt.grid()
plt.savefig("ISW ogandpartner.png")
```

```python
#Now the turning point plots(figure 2):
E = 5

linear_x = np.linspace(-0.1, 0.1, 1000)
linear_y = np.linspace(-0.1, 0.1, 1000)+5

linear_figures, (lin1, lin2) = plt.subplots(1, 2)
#linear_figures.suptitle("Left and Right Hand turning point examples")
linear_figures.set_figwidth(8)
lin1.plot(linear_x, linear_y, label = "Potential")
lin1.plot(linear_x, E*np.ones(1000),"--", label = "E")
lin1.set_xlabel("x")
lin1.set_ylabel("$V(x)$")
lin1.set_xlim(-0.1, 0.1)
lin1.set_title("Right Hand Turning Point")
lin1.legend()
lin1.grid()
lin2.plot(linear_x, linear_y[::-1], label = "Potential")
lin2.plot(linear_x, 5*np.ones(1000),"--", label = "E")
lin2.set_title("Left Hand Turning Point")
lin2.set_xlabel("x")
lin2.set_xlim(-0.1, 0.1)
lin2.legend()
lin2.grid()
plt.savefig("TurningPointExamples.png")
```

```python
#Now for the sketch of the Harmonic oscillators turning points(figure 3):

E = 4
def HarmonicOscillator(x):
    """ 
    
    Returns the harmonic oscillator potential(x^2) for the given x-values
    
    """
    return x**2

HO_line = np.linspace(-3, 3, 1000)
HO_E = np.ones(1000)*E
HO_y_values = HarmonicOscillator(HO_line)
TurningPoints = np.array([[-2, 2],[0, 0]])
Green_line_1 = np.array([[-2, -2],[0, 4]])
Green_line_2 = np.array([[2, 2],[0, 4]])

plt.plot(HO_line, HO_y_values, label="Potential")
plt.plot(HO_line, HO_E, "--", label="E")
plt.scatter(TurningPoints[0], TurningPoints[1], color="green")
plt.plot(Green_line_1[0], Green_line_1[1], "--", color="green")
plt.plot(Green_line_2[0], Green_line_2[1], "--", color="green")
plt.annotate("$x_1$", (-2, 0))
plt.annotate("$x_2$", (2, 0))
#plt.title("Harmonic Oscillator potential example, $x^2$, with energy of 4, and turning points, $x_1$ and $x_2$, at $\pm2$.")
plt.xlabel("x")
plt.ylabel("V(x)")
plt.xlim(-3, 3)
plt.ylim(-0.1, 9)
plt.grid()
plt.legend(loc="upper right")
plt.savefig("TurningPointHOexample.png")
```
